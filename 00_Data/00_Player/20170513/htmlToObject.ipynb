```python
# inport BeautifulSoup
from bs4 import BeautifulSoup
```

```python
file=open("a.html", "r+")
soup = BeautifulSoup(file.read(), "html.parser")
```

```python
print(soup.find_all("div","playerInfo"))
```

```python
for div in soup.find_all("div","playerInfo"):
    lines = div.get_text()
    lines = lines.replace(':\n                \n\n ', ':')
    lines = lines.replace('\n\n\n', '')
    lines = lines.replace('\n\n                        1', '\n1')
    lines = lines.replace('\n\n                        2', '\n2')
    lines = lines.replace('\n\n                        3', '\n3')
    lines = lines.replace('TSI','\nTSI')
    lines = lines.replace('\xa0', '')
    lines = lines.replace(', 임', '\n임')
    lines = lines.replace('Won', ' Won\n')
    lines = lines.replace('form, ', 'form\n')
    if lines.find('체력 [') > -1 :
        lines = lines.replace('체력', '체력\n특수능력')
        lines = lines.replace(']', ']\n')
    else :
        lines = lines.replace('체력', '체력\n특수능력 없음\n')
    lines = lines.replace('경험과 ', '경험\n')
    lines = lines.replace('리더십 능력', '리더십능력\n')
    lines = lines.replace('\n\n', '\n')
    lines = lines.replace('\n                    ', '')
    lines = lines.replace('골키핑', '\n골키핑')
    lines = lines.replace('수비력','\n수비력')
    lines = lines.replace('플레이메이킹', '\n플레이메이킹')
    lines = lines.replace('윙어', '\n윙어')
    lines = lines.replace('패스력','\n패스력')
    lines = lines.replace('득점력', '\n득점력')
    lines = lines.replace('세트 피스','\n세트 피스')
    print(lines)
```
